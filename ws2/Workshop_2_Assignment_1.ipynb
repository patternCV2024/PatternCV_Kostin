```
# Імпорт необхідних бібліотек
import os
import numpy as np
import tensorflow as tf


# Завантажити дані

# Отримати поточний робочий каталог
current_dir = os.getcwd()

# Додати data/mnist.npz до попереднього шляху, щоб отримати повний шлях
data_path = os.path.join(current_dir, "data/mnist.npz")

# Отримати тільки навчальний набір
(training_images, training_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()


# GRADED FUNCTION: reshape_and_normalize

def reshape_and_normalize(images):

    # Змінити форму зображень, щоб додати додатковий вимір
    images_normalized = images.reshape(-1, 28, 28, 1)

    # Нормалізація значень пікселів
    images_normalized = images_normalized / 255.0

    return images_normalized

# Перезавантажте зображення, якщо ви запускаєте цю клітинку кілька разів
(training_images, _), _ = tf.keras.datasets.mnist.load_data()

# Застосуйте свою функцію
training_images = reshape_and_normalize(training_images)

print(f"Максимальне значення пікселів після нормалізації: {np.max(training_images)}\n")
print(f"Форма навчального набору після переформування: {training_images.shape}\n")
print(f"Форма одного зображення після зміни форми: {training_images[0].shape}")

# GRADED CLASS: myCallback

# Не забувайте успадковувати від правильного класу
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if logs.get('accuracy') > 0.995:
            print("\nДосягнуто 99,5% точності, тому навчання скасовано!")
            self.model.stop_training = True

# GRADED FUNCTION: convolutional_model
def convolutional_model():

    # Визначте модель
    model = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])

    # Скомпілюйте модель
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    return model

# Збережіть вашу необучену модель
model = convolutional_model()

# Створіть екземпляр класу зворотного виклику
callbacks = myCallback()

# Навчіть свою модель (це може зайняти до 5 хвилин)
history = model.fit(training_images, training_labels, epochs=10, callbacks=[callbacks])

print(f"Ваша модель навчалася протягом {len(history.epoch)} епох")
```
