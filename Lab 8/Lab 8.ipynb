## Побудова дерева ухвалення рішень мовою python

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Функція для створення сітки
def get_grid(data):
    x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
    y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
    return np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

# Ваші дані
x_train_9 = np.array([[37, 46], [26, 50], [23, 6], [21, 33], [36, 42], [37, 44], [19, 6], [23, 5], [35, 28]])
y_train_9 = np.array([1, -1, 1, 1, -1, 1, -1, 1, 1])

# Ініціалізація та навчання моделі дерева ухвалення рішень
clf_tree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=5)
clf_tree.fit(x_train_9, y_train_9)

# Отримання прогнозів та відображення сітки та точок даних
xx, yy = get_grid(x_train_9)
predicted = clf_tree.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.pcolormesh(xx, yy, predicted, cmap='spring', shading='auto')
plt.scatter(x_train_9[:, 0], x_train_9[:, 1], c=y_train_9, s=50, cmap='spring', edgecolors='black', linewidth=1.5)
plt.show()

# Візуалізація дерева
plot_tree(clf_tree)
```

_Процес триває до тих пір, поки всі дані не будуть розділені на чисті класи або досягнута максимальна глибина дерева._
