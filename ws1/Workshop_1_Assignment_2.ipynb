```
import os
import tensorflow as tf
from tensorflow import keras

# Завантажити дані

# Отримати поточний робочий каталог
current_dir = os.getcwd()

# Додати data/mnist.npz до попереднього шляху, щоб отримати повний шлях
data_path = os.path.join(current_dir, "data/mnist.npz")

# Відмінити тестовий набір
(x_train, y_train), _ = tf.keras.datasets.mnist.load_data()

# Нормалізувати значень пікселів
x_train = x_train / 255.0

data_shape = x_train.shape

print(f"There are {data_shape[0]} examples with shape ({data_shape[1]}, {data_shape[2]})")

# GRADED CLASS: myCallback

# Запам'ятати успадкування від правильного класу
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if logs.get('accuracy') > 0.99:
            print("\nReached 99% accuracy so cancelling training!")
            self.model.stop_training = True


# GRADED FUNCTION: train_mnist
def train_mnist(x_train, y_train):

    # Створити екземпляр класу зворотного виклику
    callbacks = myCallback()

    # Визначити модель
    model = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])


    # Скомпілювати модель
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # Налаштувати модель на 10 епох, додавши зворотні виклики
    # і зберегти історію навчання
    history = model.fit(x_train, y_train, epochs=10, callbacks=[callbacks])

    return history

hist = train_mnist(x_train, y_train)
```
